```python
## file with general functions and variables useful for a lot of MSEAL stuff
import pandas as pd
import numpy as np
import os
import math
pd.set_option('max_colwidth', 100)
pd.set_option('max_rows', 10000)
```

```python
bbarrel_pdbs = pd.read_table("../DB/MonoDB_v5.txt",header=None,names=['pdb_name'])
display(bbarrel_pdbs.head(2))
print(bbarrel_pdbs.shape)
```

```python
bbarrel_df = bbarrel_pdbs.copy()
bbarrel_df['first_barrel_seqID']=""
bbarrel_df['last_barrel_seqID']=""
bbarrel_df['num_strands']=0

display(bbarrel_df[:2])
```

```python
def get_first_and_last_strand_res(pdb):
    strand_df = pd.read_csv("../Output/mono_v5/PolarBearal/betaBarrelStrands/%s.txt"%(pdb[:4]),sep='\t')
    #display(strand_df)
    first_res_seqID = strand_df.loc[0, '_seqID']
    last_res_seqID = strand_df.loc[(len(strand_df)-1), '_seqID']
    num_stands = strand_df.loc[0, '_pdb_strands']
    return(first_res_seqID, last_res_seqID, num_stands)
get_first_and_last_strand_res("4RLC_A")
```

```python
for index, row in bbarrel_df.iterrows():
    cur_first_seqID, cur_last_seqID, cur_num_stands = get_first_and_last_strand_res(row['pdb_name'])
    bbarrel_df.loc[index,'first_barrel_seqID']=cur_first_seqID
    bbarrel_df.loc[index,'last_barrel_seqID']=cur_last_seqID
    bbarrel_df.loc[index,'num_strands']=cur_num_stands
bbarrel_df['num_barrel_region_resis']=bbarrel_df['last_barrel_seqID']-bbarrel_df['first_barrel_seqID']
bbarrel_df = bbarrel_df.sort_values('num_barrel_region_resis').reset_index(drop=True)
bbarrel_df[['pdb_name', 'num_strands', 'num_barrel_region_resis']]
```
